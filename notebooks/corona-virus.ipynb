```python
import numpy as np
import pandas as pd
import os
import json
```

```python
uri = "http://i.snssdk.com/forum/ncov_data/?data_type=%5B4%5D"
data_file = "/kaggle/input/ncov_data"
```

```python
with open(data_file, 'r') as fh:
    data_obj = json.load(fh)
    overseas_data = json.loads(data_obj['overseas_data'])
    
print(json.dumps(overseas_data['country'], indent=4))
```

```python
all_series_data  = []
all_totals_data = []
all_incr_data = []

def add_common(df, c):
    for key in ("id", "code", "name"):  
            df.loc[:, key] = c[key]
            
for c in overseas_data['country']:

    df_totals = pd.DataFrame.from_records(c['countryTotal'], index=[0])
    add_common(df_totals, c)
    all_totals_data.append(df_totals)
    
    df_incr = pd.DataFrame.from_records(c['countryIncr'], index=[0])
    add_common(df_incr, c)
    all_incr_data.append(df_incr)
    
    df_series = pd.DataFrame.from_records(c['series'])
    add_common(df_series, c)
    all_series_data.append(df_series)
        
all_series = pd.concat(all_series_data, sort=False, ignore_index=True)
all_series.loc[:, 'date'] = pd.to_datetime(all_series.loc[:, 'date'])
all_series = all_series.sort_values(by=['date','id']).reset_index(drop=True)
all_totals = pd.concat(all_totals_data, sort=False, ignore_index=True)
all_incr = pd.concat(all_incr_data, sort=False, ignore_index=True)
```
