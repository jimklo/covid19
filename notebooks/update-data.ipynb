```python
import requests
import json
from pathlib import Path
from datetime import datetime
```

```python
uri = "http://i.snssdk.com/forum/ncov_data/?data_type=%5B4%5D"
data_path = Path("../data")
```

```python
res = requests.get(uri)
raw_data = res.json()
```

```python
if 'overseas_data' in raw_data and isinstance(raw_data['overseas_data'], str):
    overseas_data = json.loads(raw_data['overseas_data'])
    updateTime = datetime.utcfromtimestamp(overseas_data['updateTime'])
    raw_data_file = data_path / F"ncov_data_raw_{updateTime:%Y%m%dT%H%M%S}.json"
    clean_data_file = data_path / F"ncov_data_clean_{updateTime:%Y%m%dT%H%M%S}.json"
  
    print(F"Saving {raw_data_file}")
    with open(raw_data_file, 'w') as rd:
        json.dump(raw_data, rd)
            

    print(F"Saving {clean_data_file}")
    raw_data['overseas_data'] = overseas_data
    with open(clean_data_file, 'w') as cd:
        json.dump(raw_data, cd)
    
```
